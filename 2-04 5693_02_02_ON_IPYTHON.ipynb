## Code to solve numberically the Lorenz system

```python
import numpy
from numpy import linspace
import scipy
from scipy.integrate import odeint
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

sigma=10.0
b=8.0/3.0
r=28.0

f=lambda x,t: [sigma*(x[1]-x[0]),
               r*x[0]-x[1]-x[0]*x[2],
               x[0]*x[1]-b*x[2]]
t=linspace(0,20,2000);y0=[5.0,5.0,5.0]
solution=odeint(f,y0,t)
X=solution[:,0];Y=solution[:,1];Z=solution[:,2]
```

## Code to plot the obtained result of the Lorenz system

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = (10.0, 10.0)

theplot = plt.figure().gca(projection='3d')
theplot.plot(X,Y,Z)
theplot.set_xlabel('X')
theplot.set_ylabel('Y')
theplot.set_zlabel('Z')

plt.rc('text', usetex=False)
plt.rc('font', family='Arial')

theplot.set_title(r'Solution of the Lorenz system', fontsize=16, color='gray')
plt.show()
```
