### 3D data plot via Matplotlib

This example is based on the [Gnuplot](http://www.gnuplot.info/) [http://www.gnuplot.info/]
3D glass data set [demo](https://github.com/gnuplot/gnuplot/blob/master/demo/glass.dat) available at 
[ https://github.com/gnuplot/gnuplot/blob/master/demo/glass.dat ]

```python
x = []
y = []
z = []

filename = './Data_3D_ex_glass.dat'
with open(filename, 'r') as readingfile:
    for line in readingfile:
        data = line.strip().split()
        x.append(float(data[0]))
        y.append(float(data[1]))
        z.append(float(data[2]))
```

```python
import numpy as np
nfilas  =  8 ; ncolums =  32 

X = np.array(x).reshape(nfilas,ncolums)
Y = np.array(y).reshape(nfilas,ncolums)
Z = np.array(z).reshape(nfilas,ncolums)
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import axes3d

fig = plt.figure(figsize=(8, 6))

ax = fig.add_subplot(1,2,1, projection='3d')
ax.plot(x, y, z, 'ob-')

ax.set_xlabel('X',fontsize=16)
ax.set_ylabel('Y',fontsize=16)
ax.set_zlabel('Z',fontsize=16)
ax.set_title('A 3D view of a data set',fontsize=16)

ax = fig.add_subplot(1,2,2, projection='3d')
surf = ax.plot_surface(X, Y, Z,         
                       rstride=1,     
                       cstride=1,   
                       linewidth=1,
                       color='w'
                      )
ax.set_xlabel('X',fontsize=16)
ax.set_ylabel('Y',fontsize=16)
ax.set_zlabel('Z',fontsize=16)
ax.set_title('A 3D view of the data set surface',fontsize=16)

fig.tight_layout()
fig.savefig("Data_3D_ex_glass_fig.png")
plt.show()
```

### 3D surface plot via Matplotlib

```python
import numpy as np

x = [0.0, 1.5, 3.0, 4.5, 6.0, ]
y = [0.0, 0.86, 1.71, 2.57, 3.43, 4.29, 5.14, 6.0]

X, Y  = np.meshgrid(x,y)

Z = 4*X**2 + Y**2
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import axes3d

fig = plt.figure()
ax = fig.add_subplot(1,2,1, projection='3d')
ax.plot(X.flatten(), Y.flatten(), Z.flatten(), 'ob')
ax.plot_wireframe(X, Y, Z, rstride=1, cstride=1, color='r')

ax.set_xlabel('X',fontsize=16)
ax.set_ylabel('Y',fontsize=16)
ax.set_zlabel('Z',fontsize=16)
ax.set_title('A view of a surface',fontsize=16)

ax = fig.add_subplot(1,2,2, projection='3d')
surf = ax.plot_surface(X, Y, Z,         
                       rstride=1,     
                       cstride=1,   
                       linewidth=1,
                       color='w'
                      )

ax.set_xlabel('X',fontsize=16)
ax.set_ylabel('Y',fontsize=16)
ax.set_zlabel('Z',fontsize=16)
ax.set_title('Another view of the surface',fontsize=16)


fig.tight_layout()
fig.savefig("fig_chap_07_3Dfig_ex_2.png")
plt.show()
```
