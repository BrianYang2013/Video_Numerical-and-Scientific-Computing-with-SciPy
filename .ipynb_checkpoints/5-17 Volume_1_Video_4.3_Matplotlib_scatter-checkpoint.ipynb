### Scatter plots via  Matplotlib


```python
import numpy as np
x  = np.array([1.5, 2.7, 3.8, 9.5,12.3])
y =  np.array([3.8,-2.4, 0.35,6.2,1.5])
```

```python
import matplotlib.pyplot as plt
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(1, 1, 1)

ax.scatter(y, x, label='Scatter Plot 1')
ax.scatter(x, y, label='Scatter plot 2', s=500, c='red', marker='*')
ax.plot(2*x, 2*y, 'bo--', label='THIS Plot legend', lw=2, ms=10)

ax.set_title('THE TITLE LABEL', fontsize = 10)
ax.set_xlabel('THE HORIZONTAL (X-AXIS) LABEL', fontsize = 12)
ax.set_ylabel('THE VERTICAL (Y-AXIS) LABEL', fontsize = 15)
ax.legend(loc=4)
ax.xaxis.set_tick_params(labelsize=20)
ax.set_xticks([-4,5,14])
fig.savefig("fig0.png")
plt.show()
```

### Contour plots via  Matplotlib


```python
x = []
y = []
z = []

filename = './Data_3D_ex_glass.dat'
with open(filename, 'r') as readingfile:
    for line in readingfile:
        data = line.strip().split()
        x.append(float(data[0]))
        y.append(float(data[1]))
        z.append(float(data[2]))
```

```python
import numpy as np
nfilas  =  8 ; ncolums =  32 

X = np.array(x).reshape(nfilas,ncolums)
Y = np.array(y).reshape(nfilas,ncolums)
Z = np.array(z).reshape(nfilas,ncolums)
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import axes3d

fig = plt.figure()
ax = fig.add_subplot(1,2,1, projection='3d')
ax.scatter(X.flatten(), Y.flatten(), Z.flatten(), 'ob')

ax.set_xlabel('X',fontsize=16)
ax.set_ylabel('Y',fontsize=16)
ax.set_zlabel('Z',fontsize=16)
ax.set_title('Scatter plot of the surface',fontsize=16)

# THE CONTOUR PLOT IS GENERATED BELOW
ax = fig.add_subplot(1,2,2)
ax.contour(X, Y, Z)

ax.set_xlabel('X',fontsize=16)
ax.set_ylabel('Y',fontsize=16)
ax.set_title('Contours of the surface',fontsize=16)


fig.tight_layout()
fig.savefig("fig_chap_07_3Dfig_ex_2.png")
plt.show()
```
